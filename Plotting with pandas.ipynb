# Plotting with Pandas Library
### The Data used for this analysis was gotten from Yahoo Finance Apple.Inc Stock Data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
AAPL = pd.read_csv('AAPL.csv', index_col='Date', parse_dates=True)
AAPL.head(6)
```

#### Plotting Arrays (Matplotlib)

```python
close_array = AAPL['Close'].values
```

```python
type(close_array)
```

```python
plt.plot(close_array)
plt.show()
```

#### Plotting Series (Matplotlib)

```python
close_series = AAPL['Close']
```

```python
type(close_series)
```

```python
plt.plot(close_series)
plt.show()
```

#### Plotting Series (Pandas)

```python
close_series.plot()
plt.show()
```

#### Plotting Dataframes (Pandas)

```python
AAPL.plot()
plt.show()
```

#### Plotting Dataframes (Matplotlib)

```python
plt.plot(AAPL)
plt.show()
```

#### Fixing Scales

```python
AAPL.plot()
plt.yscale('log')
plt.show()
```

#### Customizing Plots

```python
AAPL['Open'].plot(color='b', style='.-', legend=True)
AAPL['Close'].plot(color='r', style='.', legend=True)
plt.show()
```

#### Saving Plots

```python
AAPL['Open'].plot(color='b', style='.-', legend=True)
plt.savefig('AAPL.jpg')
plt.show()
```
