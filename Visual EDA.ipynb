# Visual Exploratory Data Analysis
The Iris data set contains

* 150 observations, 4 features each

* Sepal length

* Sepal width

* Petal length

* Petal width

* 3 species: setosa, versicolor, virginica
### The Dataset was obtained from (http://archive.ics.uci.edu/ml/datasets/Iris)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Data Import

```python
iris = pd.read_csv('IRIS.csv')
iris.head()
```

```python
iris.shape
```

#### Line Plot

```python
iris.plot(x = 'sepal_length', y = 'sepal_width')
plt.show()
```

#### Scatter Plot

```python
iris.plot(x = 'sepal_length', y = 'sepal_width', kind = 'scatter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

#### Box Plot

```python
iris.plot(x = 'sepal_length', y = 'sepal_width', kind = 'box')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

#### Histogram

```python
iris.plot(x = 'sepal_length', y = 'sepal_width', kind = 'hist')
plt.xlabel('sepal length (cm)')

plt.show()
```

#### Histogram Options
* bins (integer): number of intervals or bins

* range (tuple): extrema of bins (minimum, maximum)

* normed (boolean): whether to normalize to one

* cumulative (boolean): compute Cumulative Distribution Function (CDF)

* ... more Matplotlib customizations

#### Customizing Histograms

```python
iris.plot(x = 'sepal_length', y = 'sepal_width', kind = 'hist', bins=30)
plt.xlabel('sepal length (cm)')

plt.show()
```

#### Cumulative Distribution

```python
iris.plot(x = 'sepal_length', y = 'sepal_width', kind = 'hist', bins=30, cumulative=True)
plt.xlabel('sepal length (cm)')
plt.title('Cumulative Distribution Function (CDF)')

plt.show()
```

#### Warning!!
* Three different DataFrame plot idioms

 * iris.plot(kind=hist)

 * iris.plt.hist()

 * iris.hist()

* Syntax/results differ!

* Pandas API still evolving: check documentation!

#### Computing Errors

```python
error_versicolor = 100 * np.abs(describe_versicolor - describe_all)
```

```python
error_versicolor = error_versicolor/describe_versicolor
```
